```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../assets/imori.jpg", 0)
ret, thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
edges = cv2.Canny(img, 100, 200)
plt.figure(figsize=(10,10))
plt.subplot(131),plt.imshow(img, cmap="gray")
plt.subplot(132),plt.imshow(thresh, cmap="gray")
plt.subplot(133),plt.imshow(edges, cmap="gray")
```

```python
kernel = np.array([[0, 1, 0],
                             [1, 0, 1],
                             [0, 1, 0]]).astype(np.uint8)
```

```python
# Dilation
dilation = cv2.dilate(thresh, kernel, iterations=2)
# Erosion
erosion = cv2.erode(thresh, kernel, iterations=2)
# Opening
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
# Closing
closing = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)
# Gradient
gradient = cv2.morphologyEx(thresh, cv2.MORPH_GRADIENT, kernel)
# Tophat
tophat = cv2.morphologyEx(thresh, cv2.MORPH_TOPHAT, kernel, iterations=3)
# Blackhat
blackhat = cv2.morphologyEx(thresh, cv2.MORPH_BLACKHAT, kernel, iterations=3)
```

```python
# Titles
titles = ["Dilation", "Erosion", "Opening", "Closing", "Gradient", "Tophat", "Blackhat"]
# Figures in list
figures = [dilation, erosion, opening, closing, gradient, tophat, blackhat]
```

```python
fig = plt.figure(figsize=(10,10))
for i in range(0,7):
    fig.add_subplot(3,3,i+1)
    plt.title(titles[i], fontsize=15)
    plt.tight_layout()
    plt.imshow(figures[i], cmap="gray")
```
