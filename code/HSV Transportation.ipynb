```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../assets/imori.jpg")
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
print(hsv.shape,hsv.dtype)
plt.imshow(hsv[:,:,::-1])
```

```python
hsv[:,:,0] = (hsv[:,:,0] - 180)  % 360  # (H(0~360) - 180 ) % 360 = Reverse
img2 = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
plt.subplot(121)
plt.imshow(hsv[:,:,::-1])
plt.subplot(122)
plt.imshow(img2[:,:,::-1])
```

```python
print((0 -180) % 360)  # Red
print((60 -180) % 360)  # Yellow
print((120 -180) % 360)  #  Green
print((180 -180) % 360)  #  Light Blue
print((240 -180) % 360)  #  Blue
print((300 -180) % 360)  #  Purple
print((360 -180) % 360)  #  Red
```
