```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread("../assets/imori.jpg")
img2 = cv2.imread("../assets/thorino.jpg")
```

```python
# Resize img
print(img1.shape, img2.shape)
img2 = cv2.resize(img2,(128,128))
```

```python
# Alpha Blending
dst = cv2.addWeighted(img1, 0.6, img2, 0.4, 0).astype(np.uint8)
```

```python
plt.imshow(dst[:, :, ::-1])
```
