```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../assets/imori.jpg", 0).astype(np.float32)
```

```python
pyramid = [img]
for i in range(1, 6):
    out = cv2.pyrDown(img)
    out = cv2.pyrUp(out)
    pyramid.append(out)
```

```python
h, w= pyramid[0].shape
```

```python
out = np.zeros((h, w), dtype=np.float32)
out += np.abs(pyramid[0] - pyramid[1])
out += np.abs(pyramid[0] - pyramid[3])
out += np.abs(pyramid[0] - pyramid[5])
out += np.abs(pyramid[1] - pyramid[4])
out += np.abs(pyramid[2] - pyramid[3])
out += np.abs(pyramid[3] - pyramid[5])
out = out / out.max() * 255
```

```python
plt.imshow(out, cmap="gray")
```
