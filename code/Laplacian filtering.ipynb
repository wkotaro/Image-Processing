```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Normalize
def normalize(img):
    img_min = img.min()
    img_max = img.max()
    
    out = (img - img_min) / (img_max - img_min)
    return out
```

```python
img = cv2.imread("../assets/imori.jpg", 0)
```

```python
# Laplacian filtering
lap = cv2.Laplacian(img, cv2.CV_8U, ksize=1).astype(np.uint8)
plt.imshow(lap, cmap="gray")
```

```python
# Other answer
kernel = np.array([[0, 1,  0],
                   [1, -4, 1],
                   [0, 1,  0]])

dst = cv2.filter2D(img, cv2.CV_8U, kernel)
plt.imshow(dst, cmap="gray")
```
