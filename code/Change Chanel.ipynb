```python
import cv2
from matplotlib import pyplot as plt
```

```python
img_bgr = cv2.imread("../assets/imori.jpg") # BGR
plt.imshow(img_bgr)
```

```python
img_rgb = img_bgr[:, :, ::-1] # RGB
plt.imshow(img_rgb)
```

```python
img_rgb = img_bgr[:, :, [2,1,0]] # [::-1] = [2,1,0]
plt.imshow(img_rgb)
```

```python
img_grb = img_bgr[:,:,[1,2,0]] # BGR -> GRB
img_gbr = img_bgr[:,:,[1,0,2]] # BGR -> GBR
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img_grb)
plt.subplot(122)
plt.imshow(img_gbr)
```

```python
# Slice
lst = [1, 2, 3, 4, 5, 6]
print(lst) # All
print(lst[:]) # All
print(lst[0:6]) # 0~6
print(lst[0:6:1]) # 0~6 skip 1
print(lst[0:6:2]) # 0~6 skip 2
print(lst[::-1]) # Reverse
print(lst[-1:-7:-1]) 
#[[-7,-6,-5,-4,-3,-2,-1]
#  [1 ,2   ,3  ,4  ,5  ,6]  
```
