```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def gaussian_kernel(ksize=3):
    combs = [1]
    
    for i in range(1, ksize):
        ratio = (ksize-i)/(i)
        combs.append(combs[-1]*ratio)
        
    combs = np.array(combs).reshape(1,ksize)/(2**(ksize-1))
    return combs.T.dot(combs)
```

```python
print(gaussian_kernel(3))
```

```python
img = cv2.imread("../assets/imori_noise.jpg", 0)
g = cv2.filter2D(img, cv2.CV_8U, gaussian_kernel(3))
blur = cv2.GaussianBlur(img, (3,3), 1.3)
plt.subplot(121)
plt.imshow(g, cmap="gray")
plt.subplot(122)
plt.imshow(blur, cmap="gray")
```
