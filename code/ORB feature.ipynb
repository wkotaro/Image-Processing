```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../assets/imori.jpg")
```

```python
# ORB
detector = cv2.ORB_create()
```

```python
# detecting feature
keypoints = detector.detect(img)
```

```python
img_orb = cv2.drawKeypoints(img, keypoints, None)
plt.imshow(img_orb[:,:,::-1])
```
