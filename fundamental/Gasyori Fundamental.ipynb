```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read image
img = cv2.imread("../assets/imori.jpg")
```

```python
# Show image
plt.imshow(img[:,:,::-1])# BGR -> RGB
# Other way
# cv2.imshow("imori", img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# Image Size
print(img.shape)
```

```python
# Image Type
print(img.dtype)
```

```python
# Change Type
_imt = img.astype(np.float32)
print(_img.dtype)
```

```python
# Image Max, Min, Mean
print(img.max(), img.min(), img.mean())
```

```python
# Operate image
img[20, 30] # Investigate x=20, y=30
```

```python
img[20,30,1] # investigate x=20 y=30, Color : G
```

```python
# Slice
img[20, 30:33] # Investigate x=20, y=[30,32]
# img[20, 30:] -> x=20, y=[30~last]
# img[20, :30] -> x=20, y=[start~30]
```

```python
# Copy img
img2 = img.copy()
# Turn black at x=[50], y=[:50] 
img2[:50, :50] = 0
plt.imshow(img2[:,:,::-1])
```

```python
# Copy img , Type float32
img2 = img.copy().astype(np.float32)
# x=[60:100], y=[60,100], Color : B -> 260
img2[60:100, 60:100, 0] = 260
# Return Type
plt.imshow(img2.astype(np.uint8)[:,:,::-1])  # Color : B = 260 - 256 = 4
```
