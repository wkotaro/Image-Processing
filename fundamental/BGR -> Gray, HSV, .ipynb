```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read image
img = cv2.imread("../assets/imori.jpg")
# BGR -> Gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# BGR -> HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(25,25))   # Change Figure size  plt.figure(figsize)
plt.subplot(1,3,1)   # Plot plt.subplot(Col, Row, Num)
plt.imshow(img[:, :, ::-1]) # BGR -> RGB
plt.subplot(1,3,2)
plt.imshow(gray, cmap="gray")   # Gray with matplotlib -> cmap = "gray"
plt.subplot(1,3,3)
plt.imshow(hsv[:,:,::-1]) 
```
