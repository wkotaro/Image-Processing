```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../assets/imori.jpg")
plt.imshow(img[:,:,::-1])
```

```python
# BGR -> GRAY
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Binarization (th = 128)
ret, img_th = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)  # 1: input, 2: Threshold, 3: max, 4: flag
plt.imshow(img_th, cmap="gray")
```

```python
# Otsu's binarization
ret, img_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(img_otsu, cmap="gray")
```

```python
# Plot Histgram 
plt.hist(gray.ravel(), bins=255, rwidth=0.8, range=(0, 255))  # np.ravel() = MD lst to 1D lst 
plt.xlabel('value')
plt.ylabel('appearance')
plt.show()
```
