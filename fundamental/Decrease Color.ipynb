```python
import cv2
import numpy as np
from  matplotlib import pyplot as plt
```

```python
# Decrease Color     Px = {32, 96, 160, 224} only
def decrease_color(img):
    out = img.copy()
    out = out // 64 * 64 + 32
    #val = {  32  (  0 <= val <  64)
    #            96  ( 64 <= val < 128)
    #         160  (128 <= val < 192)
    #         224  (192 <= val < 256) }
    return out
```

```python
img = cv2.imread("../assets/imori.jpg")
```

```python
out = decrease_color(img)
plt.imshow(out[:,:,::-1])
```
