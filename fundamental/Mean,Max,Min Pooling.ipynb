```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def average_pooling(img, G=8):
    out = img.copy()
    
    h, w, ch = img.shape
    Nh = int(h / G)
    Nw = int(h / G)
    
    for y in range(Nh):
        for x in range(Nw):
            for c in range(ch):
                out[G*y:G*(y+1), G*x:G*(x+1), c] = np.mean(out[G*y:G*(y+1), G*x:G*(x+1), c]).astype(np.int)
                
    return out
```

```python
def max_pooling(img, G=8):
    out = img.copy()
    
    h, w, ch = img.shape
    Nh = int(h / G)
    Nw = int(h / G)
    
    for y in range(Nh):
        for x in range(Nw):
            for c in range(ch):
                out[G*y:G*(y+1), G*x:G*(x+1), c] = np.max(out[G*y:G*(y+1), G*x:G*(x+1), c]).astype(np.int)
                
    return out
```

```python
def min_pooling(img, G=8):
    out = img.copy()
    
    h, w, ch = img.shape
    Nh = int(h / G)
    Nw = int(h / G)
    
    for y in range(Nh):
        for x in range(Nw):
            for c in range(ch):
                out[G*y:G*(y+1), G*x:G*(x+1), c] = np.min(out[G*y:G*(y+1), G*x:G*(x+1), c]).astype(np.int)
                
    return out
```

```python
img = cv2.imread("../assets/imori.jpg")
```

```python
out = average_pooling(img)
plt.imshow(out[:,:,::-1])
```

```python
out = max_pooling(img)
plt.imshow(out[:,:,::-1])
```

```python
out = min_pooling(img)
plt.imshow(out[:,:,::-1])
```
