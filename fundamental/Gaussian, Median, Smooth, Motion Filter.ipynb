```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../assets/imori_noise.jpg")
# Gaussian Filter
blur = cv2.GaussianBlur(img, (3,3), 1.3)
plt.imshow(blur[:,:,::-1])
```

```python
# Median Filter
median = cv2.medianBlur(img, 3)
plt.imshow(median[:,:,::-1])
```

```python
img2 = cv2.imread("../assets/imori.jpg")
# Smoothing Filter
smooth = cv2.blur(img2, (3,3))
plt.imshow(sm[:,:,::-1])
```

```python
# Motion Filter
## Set kernel
kernel = np.array([[1,0,0],
                                [0,1,0],
                                [0,0,1]],np.float32)/3
motion = cv2.filter2D(img2, -1, kernel)
plt.imshow(motion[:,:,::-1])
```
